## Datos Faltantes:

Lo siguiente es encontrar las categorías con datos faltantes

```python
print(df.isnull().sum())
```

Como podemos evidenciar en la tabla de arriba las variables Edad, Frecuencia_Compra, Monto_Total_Gastado y Nivel_Satisfaccion presentan datos faltantes

```python
datosTotales = 40000
edad = (2000/datosTotales) * 100
frecCompra = edad
montoTot = (1979/datosTotales) * 100
nivSat = (1983/datosTotales) * 100

print("El porcentaje que representan de datos nulos en cada una de las categorias son los siguientes:\nEdad: {}%\nFrecuencia_Compra: {}%\nMonto_Total_Gastado: {}%\nNivel_Satisfaccion: {}%".format(edad, frecCompra, montoTot, nivSat))
```

Dado que los datos presentan un porcentaje de valores nulos inferiores al 30%, como se discutió en clase, podemos considerar que las variables cuentan con una cantidad suficiente de datos para ser conservadas, por lo cual son aptas para ser tratadas.
## Datos atipicos:
Para identificar posibles datos atípicos, representaremos cada categoría mediante un diagrama de cajas y bigotes. Esto nos permitirá visualizar si algún valor se encuentra fuera del rango esperado.

```python

cat = df.select_dtypes(include=['number']).columns

# Calcular el número de filas y columnas necesarias
num_cols = 3  # Número de columnas fijas
num_rows = int(np.ceil(len(cat) / num_cols))  # Calcular filas dinámicamente

# Crear la figura con el número correcto de subgráficos
fig, axes = plt.subplots(num_rows, num_cols, figsize=(3 * num_cols, 3 * num_rows))
axes = axes.flatten()

# Dibujar diagramas de cajas y bigotes
for idx, col in enumerate(cat):
    sns.boxplot(x=df[col], ax=axes[idx])
    axes[idx].set_title(f"Box plot de {col}")

# Eliminar los ejes sobrantes (si hay menos gráficos que espacios en la cuadrícula)
for j in range(len(cat), len(axes)):  
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
